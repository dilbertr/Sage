```python
hook_height_initial=28
```

```python
hook_winch_rate=64.1 #in inches per second
```

```python
gap_left=10 #distance between the front of the robot and the elevator
```

```python
gap_right=21 #distance between the back of the robot and the elevator
```

```python
def elevator_height(time,er):
    string_unwound=time*er
    return elevator_height_initial+3*string_unwound
```

```python
def hook_height(time):
    string_unwound=time*hook_winch_rate+gap_left+gap_right
    a=string_unwound
    x=gap_left
    y=gap_right
    return sqrt((a - x - y)*(a + x - y)*(a - x + y)*(a + x + y))/(2*a)+hook_height_initial
```

```python
g=Graphics()
g += plot(hook_height(x),(x,0,1.5),color="red")
g += plot(elevator_height(x,elevator_rate),(x,0,1.5))
g += plot(82,(x,0,1.5),color="green")
g.show()
```
